# Battle Game
Using random number generator to create a mini battle game.
* User defines player's name and rounds of battles
* Computer generate the enemy's name, and attributes
    - life points are random integer between 100 and 150
    - strengths are random integer btween 0 and 50
* Impact of each attack is a random number from N(strength, 5)
* Each round ends when any player's life falls negative
* Overall results are whichever player wins the most or it is an even game.


Using random number generator, for loop, while loop, if else.

```python
import names
import time
import random
```

```python
def battlegame():
    # input parameter
    player_a=input('Choose a name for your player:')
    n=int(input('Number of battle rounds:'))
    player_b=names.get_first_name()
    # generate attributes
    player_life_a=random.randrange(100, 150)
    player_life_b=random.randrange(100, 150)
    player_strength_a=random.randrange(50)
    player_strength_b=random.randrange(50)

    time.sleep(1)
    print('''
    ====================================
    {} v. {}
    {} games
    ===================================='''.format(player_a,player_b,n))

    time.sleep(1)
    print('''
    {} attributes
    Life: {}
    Strength: {}
    -----------------------------------'''.format(player_a,player_life_a,player_strength_a))
    time.sleep(1)
    print('''
    {} attributes
    Life: {}
    Strength: {}
    ====================================
    '''.format(player_b,player_life_b,player_strength_b))

    results=[]
    for i in range(n):
        player_life_a_round=player_life_a
        player_life_b_round=player_life_b
        time.sleep(1)
        print('''
        ==========
        Round {}
        ----------'''.format(i+1))
        while True:
            if player_life_a_round>0 and player_life_b_round>0:
                cut_ab = random.gauss(player_strength_a, 5)
                cut_ba = random.gauss(player_strength_b, 5)
                player_life_a_round-=cut_ba
                player_life_b_round-=cut_ab
                time.sleep(1)
                print('''
                {} attacked {} {}
                {} life: {}
                ----------
                {} attacked {} {}
                {} life: {}
                ----------'''.format(player_a,player_b,-cut_ab,player_b,player_life_b_round
                           , player_b,player_a,-cut_ba,player_a,player_life_a_round))
            else:
                break
        if player_life_a_round<=0 and player_life_b_round<=0:
            results.append('AB')
        elif player_life_a_round<=0:
            results.append('B')
        else:
            results.append('A')

    A_Win = sum(map(lambda x : x == 'A', results))
    B_Win = sum(map(lambda x : x == 'B', results))
    if A_Win==B_Win:
        final_result = 'Even Game'
    elif A_Win>B_Win:
        final_result = player_a + ' Won!'
    else:
        final_result = player_b + ' Won!'
    
    print('''====================================
    Summary:
    {}: {} Won {} Loss
    {}: {} Won {} Loss
    {}
    '''.format(player_a,A_Win,B_Win
              , player_b,B_Win,A_Win,
              final_result))
```

```python
battlegame()
```
